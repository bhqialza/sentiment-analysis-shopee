**Import Library**

```python
import pandas as pd
import time
from google_play_scraper import Sort, reviews
```

**Scraping Review Shopee**

```python
# Mengambil review aplikasi Shopee
all_reviews = []
continuation_token = None

while len(all_reviews) < 10000:
    r, continuation_token = reviews(
        'com.shopee.id',
        lang='id',
        country='id',
        sort=Sort.NEWEST,
        count=200,
        continuation_token=continuation_token
    )

    all_reviews.extend(r)

    print(f'Total collected: {len(all_reviews)}')

    if continuation_token is None:
        break

    time.sleep(1)
```

```python
# Simpan data ke file CSV
df_reviews = pd.DataFrame(all_reviews[:10000])
df_reviews.to_csv("shopee_reviews.csv", index=False)
print("Selesai, data disimpan sebagai 'shopee_reviews.csv'")
```
